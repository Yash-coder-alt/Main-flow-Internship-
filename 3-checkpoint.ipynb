```python
###List , Tuple and Dictionary in Python
print("-----------------------------------------------------------------------------------------------------------------------------------------")
##List in Python

thislist = ["apple", "banana", "cherry"]
list2 = [1, 5, 7, 9, 3]
list3 = [True, False, False]
#printing list
print(thislist)
print(list2)
print(list3)

#printing length of list
print(len(thislist))

#printing type 
print(type(list2))

#printing element on that index value from list
print(thislist[1])

#Changing value in list
thislist[0] = "Papaya"
print(thislist)
thislist[1:3] = ["blackcurrant", "watermelon"]
print(thislist)

#Adding value in list
thislist.append("orange")
print(thislist)

#removing value from list
thislist.remove("orange")
print(thislist)

#Adding value of different list
list4 = list2 + list3
print(list4)

print("-----------------------------------------------------------------------------------------------------------------------------------------")

##Tuple in python
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")

#printing tuple
print(thistuple)

#printing length of tuple
print(len(thistuple))

#printing type
print(type(thistuple))

#printing vakue of that index value from tuple
print(thistuple[5])

#Adding new element in tuple
tuple1=("papaya")
update_thistuple= tuple(thistuple) + (tuple1,)
print(update_thistuple)

#Removing element from tuple
update_thistuple= tuple(thistuple[:5],)
print(update_thistuple)

print("-----------------------------------------------------------------------------------------------------------------------------------------")


##Dictionaries in Python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
#printing dictionary
print(thisdict)
print(thisdict["brand"])

#printing length of dictionary
print(len(thisdict))

#printing type
print(type(thisdict))

#changing values
x = thisdict.values()
print(x) #before the change
thisdict["year"] = 2020
print(x) #after the change

#adding values
thisdict["color"] = "red"
print(thisdict)

#removing value
thisdict.pop("model")
print(thisdict)

#copy dictionary
mydict = thisdict.copy()
print(mydict)
print("-----------------------------------------------------------------------------------------------------------------------------------------")

```
